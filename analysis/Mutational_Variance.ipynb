## Calculating size factors

Calculating normalization size factors by adding 1 to all values and then subtracting the normalized raw counts by 1

```python
#### CC2344 ####
CC2344_rc = pd.read_csv('/research/projects/chlamydomonas/MAexpression/analysis/raw_counts/unnormalized_CC2344_raw.txt', delimiter = '\t', index_col = 'geneid')
CC2344_rc = CC2344_rc + 1
CC2344_rc.columns = CC2344_rc.columns.str.replace('_', '-')
geomean = pd.DataFrame()

for idx in CC2344_rc.index.values:
    geomean.at[idx, 'geomean'] = stats.gmean(CC2344_rc.loc[idx])
    
for idx in CC2344_rc.index.values:
    CC2344_rc.loc[idx] = CC2344_rc.loc[idx]/geomean.at[idx, 'geomean']
CC2344_rc.loc['sizefactor+1'] = CC2344_rc.median(axis = 0)

CC2344_sf = pd.read_csv('/research/projects/chlamydomonas/MAexpression/analysis/raw_counts/CC2344_size_factors_DESeq2', delimiter =' ', header = None)
CC2344_sf.columns = ['samples', 'deseq_sizefactor']
CC2344_sf['samples'] = CC2344_sf['samples'].str.replace('.', '-').str.replace('_', '-')
CC2344_sf = pd.concat([CC2344_sf.set_index('samples'), CC2344_rc.loc['sizefactor+1']], axis = 1)

#### CC2931 ####
CC2931_rc = pd.read_csv('/research/projects/chlamydomonas/MAexpression/analysis/raw_counts/unnormalized_CC2931_raw.txt', delimiter = '\t', index_col = 'geneid')
CC2931_rc = CC2931_rc + 1 ## Adding 1 to all raw counts
CC2931_rc.columns = CC2931_rc.columns.str.replace('_', '-')
geomean = pd.DataFrame()

for idx in CC2931_rc.index.values:
    geomean.at[idx, 'geomean'] = stats.gmean(CC2931_rc.loc[idx]) ## finding the geometric mean per gene across all samples
    
for idx in CC2931_rc.index.values:
    CC2931_rc.loc[idx] = CC2931_rc.loc[idx]/geomean.at[idx, 'geomean'] ## dividing each gene by its respective geometric mean
CC2931_rc.loc['sizefactor+1'] = CC2931_rc.median(axis = 0)

CC2931_sf = pd.read_csv('/research/projects/chlamydomonas/MAexpression/analysis/raw_counts/CC2931_size_factors_DESeq2', delimiter =' ', header = None)
CC2931_sf.columns = ['samples', 'deseq_sizefactor']
CC2931_sf['samples'] = CC2931_sf['samples'].str.replace('.', '-').str.replace('_', '-')
CC2931_sf = pd.concat([CC2931_sf.set_index('samples'), CC2931_rc.loc['sizefactor+1']], axis = 1)

## Reseting index ##
CC2931_sf = CC2931_sf.reset_index()
CC2344_sf = CC2344_sf.reset_index()

## Renaming columns ##
CC2344_sf.columns = ['full_strain_id', 'deseq_sizefactor', 'sizefactor+1']
CC2931_sf.columns = ['full_strain_id', 'deseq_sizefactor', 'sizefactor+1']
```

# Null hypothesis - Simulated poisson data

We wanted to generate the null mutational variance - expression level relationship to reveal the expectation of a poisson distribution.

```python
from scipy.stats import poisson
from scipy.stats import norm

malines = ['MA1', 'MA2', 'MA3', 'MA4', 'MA5', 'MA6', 'MA7', 'MA8', 'MA9', 'MA10', 'MA11']
columns = [ i + '-rep1' for i in malines] + [ i + '-rep2' for i in malines] + [ i + '-rep3' for i in malines]

mock_genes = pd.DataFrame(columns = columns)

#### Percentile bins based on CC-2344 read count data ####
anc = pd.read_csv('/research/projects/chlamydomonas/MAexpression/analysis/raw_counts/CC2344_deseq_filtered_normalized.txt', delimiter = '\t', index_col = 'Unnamed: 0')
anc['CC2344-ANC'] = anc[['CC2344-ANC-rep1', 'CC2344-ANC-rep2', 'CC2344-ANC-rep3']].mean(axis = 1)

percentile = [np.percentile(anc['CC2344-ANC'], i) for i in range(10, 110, 10)]

#### Removing mutators ####
mut_count = pd.concat([mutations, generations], axis = 1)
mut_count.columns = ['mutations', 'generations']
mut_gen_count = mut_count.reset_index().rename(columns = {'index':'maline'})
mut_gen_count = mut_gen_count.loc[~mut_gen_count['maline'].isin(['CC2344-L1', 'CC2931-L5'])]

#### Generating simulated mutation and generation counts for mock MA lines ####
strains = np.random.choice(mut_gen_count['maline'], len(malines), replace = False)
mut_gen_count = mut_gen_count.loc[mut_gen_count['maline'].isin(strains)]
mut_gen_count['maline'] = malines
mut_gen_count = mut_gen_count.set_index('maline')

for ma in malines:
    for rep in ['-rep1', '-rep2', '-rep3']:
        genes = []
        for m in percentile:
            std = m*0.1 ## assumes that ma lines of a mean read count (m) are m*x standard deviations from the mean
            deviation = norm.rvs(loc=0, scale = std)
            genes += list(poisson.rvs(mu = m + deviation, size = 1000))
        mock_genes[ma + rep] = genes

mock_genes = mock_genes.reset_index()
mock_genes = pd.melt(mock_genes, id_vars = 'index', value_vars = mock_genes.columns[1:], var_name = 'maline', value_name = 'raw_read_counts')
mock_genes[['maline', 'reps']] = mock_genes['maline'].str.split('-', expand = True)

for idx in mock_genes.index.values:
    mock_genes.at[idx, 'mutations'] = mut_gen_count.at[mock_genes.at[idx, 'maline'], 'mutations']
    mock_genes.at[idx, 'generations'] = mut_gen_count.at[mock_genes.at[idx, 'maline'], 'generations']

mock_genes.to_csv('/research/projects/chlamydomonas/MAexpression/analysis/raw_counts/mock_data_variablestd0.4.csv', sep = '\t', index = False, header = True)
```
