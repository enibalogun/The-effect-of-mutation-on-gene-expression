## Strenkert Coexpression Network

```python
path = '/research/projects/chlamydomonas/MAexpression/'
graph = path + 'genome_info/Stenkert.PNAS.coexpression_network.gml' 
G = nx.read_gml(graph)

a = nx.get_edge_attributes(G, "weight")

################################################################################
## Adding the distance of each edge to the gml file, (distance = weight^(-1)) ##
################################################################################
for genes in a.keys():
    G.add_edge(genes[0],genes[1], distance = a[genes])
network = nx.get_edge_attributes(G, "distance")
```

```python
network_df = pd.DataFrame.from_dict(network, orient = 'index', columns = ['distance'])
network_df = network_df.loc[network_df['distance'] != 1].reset_index()
network_df['index'] = network_df['index'].astype(str)
network_df['index'] = network_df['index'].str.replace('(', '').str.replace(')', '')
network_df[['gene1', 'gene2']] = network_df['index'].str.split(',', n = 2, expand = True)
network_df['gene1'] = network_df['gene1'].str.replace("'", '')
network_df['gene2'] = network_df['gene2'].str.replace("'", '')
network_df['index'] = network_df['index'].str.replace("'", '')
network_df['distance'] = network_df['distance'].astype(float)
network_df2 = network_df

#### Isolating genes in the network that are within the v6 genome ####
all_genes = network_df['gene1'].values.tolist() + network_df['gene2'].values.tolist()
all_genes = pd.Series(all_genes).drop_duplicates().tolist() ### all the genes in the network
annotation_file = pd.read_csv('/research/projects/chlamydomonas/MAexpression/genome_info/v5.3_to_v5.6_to_v6.csv', delimiter = '\t')
v6_genes = annotation_file.loc[annotation_file['v5.5_gene'].isin(all_genes)]['v5.5_gene'].values.tolist()
network_df = network_df.loc[(network_df['gene1'].isin(v6_genes)) | (network_df['gene2'].isin(v6_genes))]

network_df.to_csv('/research/projects/chlamydomonas/MAexpression/genome_info/Strenkert_coexpression_network/network.csv', sep = '\t', index = False, header = True)
```

### Hub genes based on intramodular connectivity

Took the genes at the top 5% intramodular connectivity of each module

```python
hub_genes = {}
for module in genesmodule.columns:
    a = IM_scaled.loc[genesmodule[module].dropna()]
    genes_5perc = a.sort_values('kWithin', ascending = False)['kWithin'][:(round(len(a)*0.05))]
    genes_5perc = genes_5perc.index.values.tolist()
    hub_genes[module] = genes_5perc
    
hub_genes = pd.DataFrame(dict([(k,pd.Series(v)) for k, v in hub_genes.items()])).drop(columns = ['grey'])

#### Creating a list of all hub genes ####
hubgeneslist = []
for module in hub_genes.columns:
    hubgeneslist += list(hub_genes[module].dropna())

#### Reformatted gene id to ascertain the enriched GO terms of the list of hub genes ####
hub_genes_reformatted = hub_genes.copy()

for module in hub_genes_reformatted.columns:
    hub_genes_reformatted[module] = hub_genes[module].str.replace('Cre','CHLRE_').str.replace('.g', 'g') + 'v5'

for module in hub_genes_reformatted.columns:
    csv = hub_genes_reformatted[module].dropna()
    csv.to_csv('/research/projects/chlamydomonas/MAexpression/genome_info/Strenkert_coexpression_network/modules/hub_genes_' + module + '.csv', sep = '\n', index = False)
    
#### Converting hub gene IDs in v5 to v ####
hubgenes_v6 = annotation_file.loc[annotation_file['v5.5_gene'].isin(hubgeneslist)]['v6'].dropna().values.tolist()
```

## Hub Gene Analysis

### Contrasting total connectivity of Hub genes vs Nonhub genes

```python
#### Difference in total connectivity of hub genes vs nonhub genes for each module ####
for module in genesmodule.columns[:-1]:
    hubgeneslist = hub_genes[module].dropna()
    hubgenes = IM_scaled.loc[hubgeneslist]['kTotal']
    complete_module = IM_scaled.loc[genesmodule[module].dropna()]['kTotal']
    nonhubgenes = complete_module.drop(hubgeneslist)
    print(module + ' ' + str(brunnermunzel(hubgenes, nonhubgenes, alternative = 'greater')))
    
#### Difference in total connectivity of all hub genes vs nonhub genes ####
brunnermunzel(IM_scaled.loc[hubgeneslist]['kTotal'], IM_scaled.drop(hubgeneslist)['kTotal'].dropna(), alternative = 'greater')
```

### Contrasting intradmodular connectivity of Hub genes vs Nonhub genes

```python
#### Difference in intramodular connectivity between hub genes and nonhub genes in each module ####
for module in genesmodule.columns[:-1]:
    hubgeneslist = hub_genes[module].dropna()
    hubgenes = IM_scaled.loc[hubgeneslist]['kWithin']
    complete_module = IM_scaled.loc[genesmodule[module].dropna()]['kWithin']
    nonhubgenes = complete_module.drop(hubgeneslist)
    print(module + ' ' + str(mannwhitneyu(hubgenes, nonhubgenes, alternative = 'greater'))) ## different variances, should use brunner munzel

#### Difference in intramodular connectivity between all hub genes and nonhub genes ####
brunnermunzel(IM_scaled.loc[hubgeneslist]['kWithin'], IM_scaled.drop(hubgeneslist)['kWithin'].dropna(), alternative = 'greater')
```

### Contrasting Expression Change in Putative Hub Genes vs Non-hub genes

### DEG Enrichment in Each Module

```python
total = 0
for module in genesmodule.columns:
    total += len(genesmodule[module].dropna())
    
expected = {}
for maline in degs_v5.columns:
    exp = []
    for module in genesmodule.columns:
        exp.append(round((len(genesmodule[module].dropna())/total)*len(degs_v5[maline].dropna()), 14))
    expected[maline] = exp

observed = {}
for maline in degs_v5.columns:
    obs = []
    for module in genesmodule.columns:
        obs.append(int(len(set(genesmodule[module].dropna()).intersection(set(degs_v5[maline].dropna())))))
    observed[maline] = obs
    
#### Enrichment via chi square analysis (Can you hypergeometric equation) ####
for maline in degs_v5.columns:
    print(scipy.stats.chisquare(observed[maline], f_exp = expected[maline]))
```
