## Plots of the total absolute sum of log<sub>2</sub> fold change per mutation accumulation line

```python
CC2344_l2f = pd.read_csv('/research/projects/chlamydomonas/MAexpression/analysis/raw_counts/CC2344_log2Fold', delimiter = '\t', index_col = 'Unnamed: 0')
CC2931_l2f = pd.read_csv('/research/projects/chlamydomonas/MAexpression/analysis/raw_counts/CC2931_log2Fold', delimiter = '\t', index_col = 'Unnamed: 0')

sum_l2f = pd.DataFrame()
for i in CC2344[1:]:
    sum_l2f.at[i, 'sum_l2f'] = CC2344_l2f[i].sum()
    sum_l2f.at[i, 'absolute_sum_l2f'] = abs(CC2344_l2f[i]).sum()
    sum_l2f.at[i, 'mutation_count'] = dic_mut[i]
    sum_l2f.at[i, 'strain'] = 'CC-2344'
for i in CC2931[1:]:
    sum_l2f.at[i, 'sum_l2f'] = CC2931_l2f[i].sum()
    sum_l2f.at[i, 'absolute_sum_l2f'] = abs(CC2931_l2f[i]).sum()
    sum_l2f.at[i, 'mutation_count'] = dic_mut[i]
    sum_l2f.at[i, 'strain'] = 'CC-2931'
    
#### Plot of the total sum of log2 fold change per MA line ####
sum_l2f = sum_l2f.reset_index()
plt.figure(figsize = (10,10))
sns.barplot(data = sum_l2f, y = 'index', x = 'sum_l2f')
plt.xlabel('Sum of log\u2082 (fold change)', fontsize = 15)
plt.ylabel('Mutation accumulation lines', fontsize = 15)
plt.savefig('/research/projects/chlamydomonas/MAexpression/analysis/log2fold/total_sum_l2f_per_MA.pdf', format = 'pdf', dpi = 600, bbox_inches = 'tight')

#### Plot of the sum of absolute log2 fold change per MA line ####
plt.figure(figsize = (10,10))
sns.barplot(data = sum_l2f, y = 'index', x = 'absolute_sum_l2f')
plt.xlabel('Sum of absolute log\u2082 (fold change)', fontsize = 15)
plt.ylabel('MA lines', fontsize = 15)
plt.savefig('/research/projects/chlamydomonas/MAexpression/analysis/log2fold/absolute_sum_l2f_per_MA.pdf', format = 'pdf', dpi = 600, bbox_inches = 'tight')

#### Scatterplot of total absolute log2 fold change per MA line ####
plt.figure(figsize = (10, 10))
a = sns.scatterplot(data = sum_l2f, x = 'mutation_count', y = 'absolute_sum_l2f', hue = 'strain')
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left', title = 'Strain', fontsize = 20)
plt.setp(a.get_legend().get_title(), fontsize='20')
plt.xlabel('Mutation count', fontsize = 30)
plt.ylabel('Sum of absolute log\u2082 (fold change)', fontsize = 30)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.savefig('/research/projects/chlamydomonas/MAexpression/analysis/log2fold/absolute_sum_l2f_per_mut.pdf', format = 'pdf', dpi = 600, bbox_inches = 'tight')

#### Scatterplot of total absolute log2 fold change per MA line (excluding mutators) ####
plt.figure(figsize = (10, 10))
a = sns.scatterplot(data = sum_l2f, x = 'mutation_count', y = 'absolute_sum_l2f', hue = 'strain')
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left', title = 'Strain', fontsize = 20)
plt.setp(a.get_legend().get_title(), fontsize='20')
plt.xlabel('Mutation count', fontsize = 20)
plt.ylabel('Sum of absolute log\u2082 (fold change)', fontsize = 20)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlim(0, 150)

### OLS regression results
results = ols('absolute_sum_l2f ~ C(strain) + mutation_count', data = sum_l2f).fit()
results.summary()
```

# FPKM based analysis

FPKM stands for fragments per kilobase per million mapped fragments. It was used as a method to compare the expression between samples in some of our analyses.

## FPKM difference vs mutation count (DEGs only)

To account for low expression genes that would have larger expression changes but in reality the change is small, we calculated the FPKM difference of a genes read count in each MA line with respect to their unmutated ancestor.

```python
CC2344 = ["CC2344-L1", "CC2344-L2", "CC2344-L3", "CC2344-L4", "CC2344-L5", "CC2344-L6", "CC2344-L7", "CC2344-L8", "CC2344-L9", "CC2344-L10", "CC2344-L11", "CC2344-L12", "CC2344-L13", "CC2344-L14", "CC2344-L15"]
CC2931 = ["CC2931-L1", "CC2931-L2", "CC2931-L3", "CC2931-L4", "CC2931-L5", "CC2931-L6", "CC2931-L7", "CC2931-L9", "CC2931-L10", "CC2931-L11", "CC2931-L13", "CC2931-L14", "CC2931-L15"]
total = CC2344 + CC2931

CC2931_avg_fpkm = pd.read_csv('/research/projects/chlamydomonas/MAexpression/analysis/fpkm/CC2931_avg_fpkm', index_col = 'index', delimiter = '\t')
CC2344_avg_fpkm = pd.read_csv('/research/projects/chlamydomonas/MAexpression/analysis/fpkm/CC2344_avg_fpkm', index_col = 'index', delimiter = '\t')

fitness = pd.read_csv('/research/projects/chlamydomonas/MAexpression/genome_info/mutation_info/Mutation_Fitness.txt', delimiter = '\t')
fitness = fitness.replace('_', '-L', regex = True)
fitness = fitness.loc[fitness['Sample'].isin(CC2344 + CC2931)].rename(columns = {'Average_realtive':'Average_relative'}).set_index('Sample')

DEGs = pd.read_csv('/research/projects/chlamydomonas/MAexpression/analysis/DEGs/total_genes1.csv', delimiter = ',')
DEGs = DEGs.reset_index()
DEGs = pd.melt(DEGs, id_vars = 'index', value_vars = DEGs.columns[1:])
DEGs = DEGs.dropna().drop(columns = ['index'])
DEGs['fpkm_difference'] = np.nan
DEGs['sample'] = DEGs['variable'].str.split('-', n = 1, expand = True)[0]

CC2344_DEGs = DEGs.loc[DEGs['variable'].isin(CC2344)]
CC2931_DEGs = DEGs.loc[DEGs['variable'].isin(CC2931)]

#### FPKM difference in expression count with respect to the ancestor ####
for i in list(CC2344_DEGs.index.values):
    DEGs.at[i, 'fpkm_difference'] = CC2344_avg_fpkm.loc[CC2344_DEGs.at[i, 'value'], CC2344_DEGs.at[i, 'variable']] - CC2344_avg_fpkm.loc[CC2344_DEGs.at[i, 'value'], 'CC2344-ANC']
for i in list(CC2931_DEGs.index.values):
    DEGs.at[i, 'fpkm_difference'] = CC2931_avg_fpkm.loc[CC2931_DEGs.at[i, 'value'], CC2931_DEGs.at[i, 'variable']] - CC2931_avg_fpkm.loc[CC2931_DEGs.at[i, 'value'], 'CC2931-ANC']
    
DEG_fpkm_sum = pd.DataFrame(index = total)
for i in CC2344 + CC2931:
    if 'CC2344' in i:
        positive_sum = DEGs.loc[DEGs['variable'] == i].loc[DEGs['fpkm_difference'] > 0]['fpkm_difference'].sum()
        negative_sum = DEGs.loc[DEGs['variable'] == i].loc[DEGs['fpkm_difference'] < 0]['fpkm_difference'].sum()
        DEG_fpkm_sum.at[i, 'positive_sum'] = positive_sum
        DEG_fpkm_sum.at[i, 'negative_sum'] = negative_sum
        DEG_fpkm_sum.at[i, 'positive_l2f_sum'] = np.log2(positive_sum)
        DEG_fpkm_sum.at[i, 'negative_l2f_sum'] = (-1)* np.log2(abs(negative_sum))
        DEG_fpkm_sum.at[i, 'generation'] = dic_gen[i]
        DEG_fpkm_sum.at[i, 'mutation'] = dic_mut[i]
        DEG_fpkm_sum.at[i, 'fitness'] = fitness.at[i, 'Average_relative']
        DEG_fpkm_sum.at[i, 'sample'] = 'CC-2344'
    if 'CC2931' in i:
        positive_sum = DEGs.loc[DEGs['variable'] == i].loc[DEGs['fpkm_difference'] > 0]['fpkm_difference'].sum()
        negative_sum = DEGs.loc[DEGs['variable'] == i].loc[DEGs['fpkm_difference'] < 0]['fpkm_difference'].sum()
        DEG_fpkm_sum.at[i, 'positive_sum'] = positive_sum
        DEG_fpkm_sum.at[i, 'negative_sum'] = negative_sum
        DEG_fpkm_sum.at[i, 'positive_l2f_sum'] = np.log2(positive_sum)
        DEG_fpkm_sum.at[i, 'negative_l2f_sum'] = (-1)* np.log2(abs(negative_sum))
        DEG_fpkm_sum.at[i, 'generation'] = dic_gen[i]
        DEG_fpkm_sum.at[i, 'mutation'] = dic_mut[i]
        DEG_fpkm_sum.at[i, 'fitness'] = fitness.at[i, 'Average_relative']
        DEG_fpkm_sum.at[i, 'sample'] = 'CC-2931'

mean_summary = DEG_fpkm_sum.mean(axis = 0)

#### Figures ####
#################
plt.figure(figsize = (10,8))
sns.scatterplot(data = DEG_fpkm_sum, x = 'mutation', y = 'positive_sum', hue = 'sample', legend = False)
plt.ylabel('Sum of FPKM differences', fontsize = 25)
slope = mean_summary['positive_sum']/mean_summary['mutation']
plt.text(300, slope*300 + 150, 'm = ' + str(slope.round(2)), fontsize = 15)
plt.plot([0, 500], [0, slope*500], color = 'lightgrey', linewidth = 4, linestyle = 'dashed')
a = sns.scatterplot(data = DEG_fpkm_sum, x = 'mutation', y = 'negative_sum', hue = 'sample')
plt.xlabel('Mutation count', fontsize = 25)

slope = mean_summary['negative_sum']/mean_summary['mutation']
plt.text(300, slope*300 + 50, 'm = ' + str(slope.round(2)), fontsize = 15)
plt.plot([0, 500], [0, slope*500], color = 'lightgrey', linewidth = 4, linestyle = 'dashed')
plt.xlim(0, 500)
plt.axhline(0, color = 'black')
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
plt.legend(bbox_to_anchor=(0.754, 1.0), loc='upper left', title = 'Strain', fontsize = 15)
plt.setp(a.get_legend().get_title(), fontsize='15')

#### Log2 transformed figure ####
plt.figure(figsize = (10,8))
sns.scatterplot(data = DEG_fpkm_sum, x = 'mutation', y = 'positive_l2f_sum', hue = 'sample', legend = False)
plt.ylabel('Sum of FPKM difference (log\u2082)', fontsize = 25)
slope = mean_summary['positive_l2f_sum']/mean_summary['mutation']
plt.text(200, slope*200 + 5, 'm = ' + str(slope.round(2)), fontsize = 15)
plt.plot([0, 400], [0, slope*400], color = 'lightgrey', linewidth = 4, linestyle = 'dashed')
a = sns.scatterplot(data = DEG_fpkm_sum, x = 'mutation', y = 'negative_l2f_sum', hue = 'sample')
plt.xlabel('Mutation count', fontsize = 25)

slope = mean_summary['negative_l2f_sum']/mean_summary['mutation']
plt.text(200, slope*200 + 5, 'm = ' + str(slope.round(2)), fontsize = 15)
plt.plot([0, 400], [0, slope*400], color = 'lightgrey', linewidth = 4, linestyle = 'dashed')
plt.axhline(0, color = 'black')
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.legend(bbox_to_anchor=(0.754, 1.0), loc='upper left', title = 'Strain', fontsize = 15)
plt.setp(a.get_legend().get_title(), fontsize='15')
plt.savefig('/research/projects/chlamydomonas/MAexpression/analysis/log2fold/fpkm_difference_vs_mut_count.eps', format = 'eps', dpi = 600, bbox_inches = 'tight')
plt.savefig('/research/projects/chlamydomonas/MAexpression/analysis/log2fold/fpkm_difference_vs_mut_count.pdf', format = 'pdf', dpi = 600, bbox_inches = 'tight')
```

```python
#### FPKM-based percentile bins ####
CC2344_bins = pd.read_csv('/research/projects/chlamydomonas/MAexpression/analysis/fpkm/CC2344_bins', delimiter = '\t')
CC2931_bins = pd.read_csv('/research/projects/chlamydomonas/MAexpression/analysis/fpkm/CC2931_bins', delimiter = '\t')

CC2344_bins = CC2344_bins.melt(id_vars = 'Unnamed: 0', value_vars = CC2344_bins.columns[1:]).dropna().set_index('value')
CC2931_bins = CC2931_bins.melt(id_vars = 'Unnamed: 0', value_vars = CC2931_bins.columns[1:]).dropna().set_index('value')

#### Conversion of metabolic genes to v6 (n = 1303) ####
metabolic_genes = pd.read_csv('/research/projects/chlamydomonas/MAexpression/analysis/divergence/iCre1355_genes.csv', delimiter = '\t')
metabolic_genes.columns = ['v5.5', 'v5.5_gene']
conversion = pd.read_csv('/research/projects/chlamydomonas/MAexpression/genome_info/v5.3_to_v5.6_to_v6.csv', delimiter = '\t')
metabolic_genes = pd.merge(metabolic_genes, conversion, on = 'v5.5')

#### Identifying genes as metabolic ####
CC2344_Vm = pd.read_csv('/research/projects/chlamydomonas/MAexpression/analysis/divergence/CC2344_L2F.csv', delimiter = '\t', index_col = 'Unnamed: 0')
CC2931_Vm = pd.read_csv('/research/projects/chlamydomonas/MAexpression/analysis/divergence/CC2931_L2F.csv', delimiter = '\t', index_col = 'Unnamed: 0')
CC2344_Vm['metabolic_gene'] = 'nonmetabolic'
CC2931_Vm['metabolic_gene'] = 'nonmetabolic'

metabolic_genes = set(CC2931_Vm.index.values).intersection(set(metabolic_genes['v6'])) ##isolating metabolic genes that weren't removed from analysis due to low read count or no longer existed in v6

CC2344_Vm.loc[list(metabolic_genes)] = CC2344_Vm.loc[list(metabolic_genes)].replace('nonmetabolic', 'metabolic')
CC2931_Vm.loc[list(metabolic_genes)] = CC2931_Vm.loc[list(metabolic_genes)].replace('nonmetabolic', 'metabolic')

#### Mutational Heritability information ####
CC2931_hert = pd.read_csv('/research/projects/chlamydomonas/MAexpression/analysis/Vm/CC2931_heritability_Vm_Ve.txt', delimiter = '\t', index_col = 'Unnamed: 0')[['Vm_gen_noint', 'Vm_mut_noint_nomutator']]
CC2931_hert.columns = ['herit_Vm_gen', 'herit_Vm_mut']

CC2344_hert = pd.read_csv('/research/projects/chlamydomonas/MAexpression/analysis/Vm/CC2344_heritability_Vm_Ve.txt', delimiter = '\t', index_col = 'Unnamed: 0')[['Vm_gen_noint', 'Vm_mut_noint_nomutator']]
CC2344_hert.columns = ['herit_Vm_gen', 'herit_Vm_mut']

## File containing metabolic genes from Cre1355 data ##
centrality = pd.read_csv('/research/projects/chlamydomonas/MAexpression/analysis/divergence/files/network_divergence.csv', delimiter = '\t', index_col = 'v6')
centrality['Ka_Ks'] = centrality['k0']/centrality['k4']

for i in ['SumDegree', 'InDegree', 'OutDegree']:
    plt.figure()
    sns.scatterplot(data = centrality, x = i, y = 'BetweennessCentrality')

centrality = centrality[~centrality.index.duplicated(keep = 'first')] ## drops v5 genes that are no longer part of v6

CC2344_gene_network = CC2344_Vm.join(centrality[['BetweennessCentrality','SumDegree', 'InDegree', 'OutDegree']])
CC2931_gene_network = CC2931_Vm.join(centrality[['BetweennessCentrality','SumDegree', 'InDegree', 'OutDegree']])

CC2344_gene_network = CC2344_gene_network.join(CC2344_hert)
CC2931_gene_network = CC2931_gene_network.join(CC2931_hert)

#### Taking the median (absolute) Log2 fold change per gene across all MA lines ####
for idx in CC2344_gene_network.index.unique():
    CC2344_gene_network.loc[idx,'average_abs_l2f'] = CC2344_gene_network.loc[idx]['abs_l2f'].median()

for idx in CC2931_gene_network.index.unique():
    CC2931_gene_network.loc[idx,'average_abs_l2f'] = CC2931_gene_network.loc[idx]['abs_l2f'].median()
    
for idx in CC2344_gene_network.index.unique():
    CC2344_gene_network.loc[idx,'average_l2f'] = CC2344_gene_network.loc[idx]['l2f'].median()

for idx in CC2931_gene_network.index.unique():
    CC2931_gene_network.loc[idx,'average_l2f'] = CC2931_gene_network.loc[idx]['l2f'].median()

#### Adding the FPKM-based percentile bin for each gene ####
CC2344_gene_network = CC2344_gene_network.join(CC2344_bins['variable'])
CC2931_gene_network = CC2931_gene_network.join(CC2931_bins['variable'])

#### Adding ancestral read counts ####
CC2344_reads = pd.read_csv('/research/projects/chlamydomonas/MAexpression/analysis/raw_counts/CC2344_deseq_filtered_normalized.txt', delimiter = '\t', index_col = 'Unnamed: 0')
CC2344_reads['CC2344-ANC'] = CC2344_reads[['CC2344-ANC-rep1', 'CC2344-ANC-rep2', 'CC2344-ANC-rep3']].mean(axis = 1)

CC2931_reads = pd.read_csv('/research/projects/chlamydomonas/MAexpression/analysis/raw_counts/CC2931_deseq_filtered_normalized.txt', delimiter = '\t', index_col = 'Unnamed: 0')
CC2931_reads['CC2931-ANC'] = CC2931_reads[['CC2931-ANC-rep1', 'CC2931-ANC-rep2', 'CC2931-ANC-rep3']].mean(axis = 1)

CC2344_gene_network = CC2344_gene_network.join(CC2344_reads['CC2344-ANC'])
CC2931_gene_network = CC2931_gene_network.join(CC2931_reads['CC2931-ANC'])

#### Viewing connections between genes ####
for i in ['SumDegree', 'InDegree', 'OutDegree', 'BetweennessCentrality']:
    plt.figure()
    sns.scatterplot(data = centrality, y = 'Ka_Ks', x = i)
```

## Metabolic genes - Expression change vs. degree centrality

We categorize the genes within each degree centrality bin according to the direction of expression change. The distribution of upregulated and down-regulated gene were plotted separately against their respective degree centrality bin. The Mann Whitney U or Brunner Munzel test was used to ascertain the presence of significant differences in median log2 fold values.

### Prepping files

```python
#### Dividing the centrality measurements into groups based on percentiles ####
bc_perc_bins = [np.percentile(CC2344_gene_network['BetweennessCentrality'].dropna(), perc) for perc in range(0, 110, 10)][1:]
bc_perc_bins[0] = -1
id_perc_bins = [np.percentile(CC2344_gene_network['InDegree'].dropna(), perc) for perc in range(0, 110, 10)]
id_perc_bins[0] = -1
od_perc_bins = [np.percentile(CC2344_gene_network['OutDegree'].dropna(), perc) for perc in range(0, 110, 10)]
od_perc_bins[0] = -1
sd_perc_bins = [np.percentile(CC2344_gene_network['SumDegree'].dropna(), perc) for perc in range(0, 110, 10)]
sd_perc_bins[0] = -1

CC2344_gene_network['BC_Groups'] = pd.cut(CC2344_gene_network['BetweennessCentrality'], bins = bc_perc_bins)
CC2931_gene_network['BC_Groups'] = pd.cut(CC2931_gene_network['BetweennessCentrality'], bins = bc_perc_bins)

CC2344_gene_network['ID_Groups'] = pd.cut(CC2344_gene_network['InDegree'], bins = id_perc_bins)
CC2931_gene_network['ID_Groups'] = pd.cut(CC2931_gene_network['InDegree'], bins = id_perc_bins)

CC2344_gene_network['OD_Groups'] = pd.cut(CC2344_gene_network['OutDegree'], bins = od_perc_bins)
CC2931_gene_network['OD_Groups'] = pd.cut(CC2931_gene_network['OutDegree'], bins = od_perc_bins)

CC2344_gene_network['SD_Groups'] = pd.cut(CC2344_gene_network['SumDegree'], bins = sd_perc_bins)
CC2931_gene_network['SD_Groups'] = pd.cut(CC2931_gene_network['SumDegree'], bins = sd_perc_bins)

#### Ordering the bins in each degree/centrality group ####
bc_ordered_bins = pd.cut(bc_perc_bins, bins = bc_perc_bins)[1:]
id_ordered_bins = pd.cut(id_perc_bins, bins = id_perc_bins)[1:]
od_ordered_bins = pd.cut(od_perc_bins, bins = od_perc_bins)[1:]
sd_ordered_bins = pd.cut(sd_perc_bins, bins = sd_perc_bins)[1:]

ordered_list = {'OD_Groups':od_ordered_bins, 'ID_Groups':id_ordered_bins, 'BC_Groups':bc_ordered_bins, 'SD_Groups':sd_ordered_bins}
```

### Absolute expression change vs degree centrality

```python
CC2344_gene_network = CC2344_gene_network[~CC2344_gene_network.index.duplicated(keep='first')]
CC2931_gene_network = CC2931_gene_network[~CC2931_gene_network.index.duplicated(keep='first')]

#### Plotting expression change against degree centrality ####
for i in ['BC_Groups', 'ID_Groups', 'OD_Groups', 'SD_Groups']:
    plt.figure(figsize = (10,10))
    sns.violinplot(data = CC2344_gene_network, x = i, y = 'average_abs_l2f')
    plt.xticks(rotation = 45)
    plt.title('CC-2344 absolute expression change')
    
for i in ['BC_Groups', 'ID_Groups', 'OD_Groups', 'SD_Groups']:
    plt.figure(figsize = (10,10))
    sns.violinplot(data = CC2931_gene_network, x = i, y = 'average_abs_l2f')
    plt.xticks(rotation = 45)
    plt.title('CC-2931 absolute expression change')
    
#### Variance and median of expression change over centrality groups ####
vm = 'average_abs_l2f'
for i in ['BC_Groups', 'ID_Groups', 'OD_Groups', 'SD_Groups']:
    stats = pd.DataFrame()
    for a in ordered_list[i]:
        stats.at[a, 'var_' + i + '_' + vm] = CC2344_gene_network.loc[CC2344_gene_network[i] == a][vm].var()
        stats.at[a, 'med_' + i + '_' + vm] = CC2344_gene_network.loc[CC2344_gene_network[i] == a][vm].median()
    display(stats)
    
    stats = pd.DataFrame()
    for a in ordered_list[i]:
        stats.at[a, 'var_' + i + '_' + vm] = CC2931_gene_network.loc[CC2931_gene_network[i] == a][vm].var()
        stats.at[a, 'med_' + i + '_' + vm] = CC2931_gene_network.loc[CC2931_gene_network[i] == a][vm].median()
    display(stats)

#### Brunner Munzel test - testing for significant differences in medians ####
bm_results = pd.DataFrame()
bm_results.at['BC_Groups', 'CC-2344_' + vm] = brunnermunzel(CC2344_gene_network.loc[CC2344_gene_network['BC_Groups'] == bc_ordered_bins[0]][vm], CC2344_gene_network.loc[CC2344_gene_network['BC_Groups'] == bc_ordered_bins[-1]][vm], alternative = 'less')[1]
bm_results.at['BC_Groups', 'CC-2931_' + vm] = brunnermunzel(CC2931_gene_network.loc[CC2931_gene_network['BC_Groups'] == bc_ordered_bins[0]][vm], CC2931_gene_network.loc[CC2931_gene_network['BC_Groups'] == bc_ordered_bins[-1]][vm], alternative = 'less')[1]
bm_results.at['ID_Groups', 'CC-2344_' + vm] = brunnermunzel(CC2344_gene_network.loc[CC2344_gene_network['ID_Groups'] == id_ordered_bins[0]][vm], CC2344_gene_network.loc[CC2344_gene_network['ID_Groups'] == id_ordered_bins[-1]][vm], alternative = 'less')[1]
bm_results.at['ID_Groups', 'CC-2931_' + vm] = brunnermunzel(CC2931_gene_network.loc[CC2931_gene_network['ID_Groups'] == id_ordered_bins[0]][vm], CC2931_gene_network.loc[CC2931_gene_network['ID_Groups'] == id_ordered_bins[-1]][vm], alternative = 'less')[1]
bm_results.at['OD_Groups', 'CC-2344_' + vm] = brunnermunzel(CC2344_gene_network.loc[CC2344_gene_network['OD_Groups'] == od_ordered_bins[0]][vm], CC2344_gene_network.loc[CC2344_gene_network['OD_Groups'] == od_ordered_bins[-1]][vm], alternative = 'less')[1]
bm_results.at['OD_Groups', 'CC-2931_' + vm] = brunnermunzel(CC2931_gene_network.loc[CC2931_gene_network['OD_Groups'] == od_ordered_bins[0]][vm], CC2931_gene_network.loc[CC2931_gene_network['OD_Groups'] == od_ordered_bins[-1]][vm], alternative = 'less')[1]
bm_results.at['SD_Groups', 'CC-2344_' + vm] = brunnermunzel(CC2344_gene_network.loc[CC2344_gene_network['SD_Groups'] == sd_ordered_bins[0]][vm], CC2344_gene_network.loc[CC2344_gene_network['SD_Groups'] == sd_ordered_bins[-1]][vm], alternative = 'less')[1]
bm_results.at['SD_Groups', 'CC-2931_' + vm] = brunnermunzel(CC2931_gene_network.loc[CC2931_gene_network['SD_Groups'] == sd_ordered_bins[0]][vm], CC2931_gene_network.loc[CC2931_gene_network['SD_Groups'] == sd_ordered_bins[-1]][vm], alternative = 'less')[1]
display(bm_results)
```

```python
#### FPKM-normalized read counts of metabolic genes ####
CC2344_fpkm = pd.read_csv('/research/projects/chlamydomonas/MAexpression/analysis/fpkm/CC2344_avg_fpkm', delimiter = '\t', index_col = 'index')
CC2344_fpkm = CC2344_fpkm.loc[metabolic_genes.index]
CC2931_fpkm = pd.read_csv('/research/projects/chlamydomonas/MAexpression/analysis/fpkm/CC2931_avg_fpkm', delimiter = '\t', index_col = 'index')
CC2931_fpkm = CC2931_fpkm.loc[metabolic_genes.index]

#### Metabolic genes with the highest and lowest expression level ####
CC2344_low_exp = CC2344_fpkm.loc[CC2344_fpkm['CC2344-ANC'] <= np.percentile(CC2344_fpkm, 10)].index
CC2344_hi_exp = CC2344_fpkm.loc[CC2344_fpkm['CC2344-ANC'] > np.percentile(CC2344_fpkm, 90)].index
CC2931_low_exp = CC2931_fpkm.loc[CC2931_fpkm['CC2931-ANC'] <= np.percentile(CC2931_fpkm, 10)].index
CC2931_hi_exp = CC2931_fpkm.loc[CC2931_fpkm['CC2931-ANC'] > np.percentile(CC2931_fpkm, 90)].index

#### Assigning high and low expression metabolic genes ####
column = ['SumDegree', 'InDegree', 'OutDegree', 'BetweennessCentrality']

CC2344_low = CC2344_gene_network[column].loc[CC2344_low_exp]
CC2344_low['expression'] = 'low'
CC2344_low['sample'] = 'CC-2344'
CC2344_high = CC2344_gene_network[column].loc[CC2344_hi_exp]
CC2344_high['expression'] = 'high'
CC2344_high['sample'] = 'CC-2344'
CC2931_low = CC2931_gene_network[column].loc[CC2931_low_exp]
CC2931_low['expression'] = 'low'
CC2931_low['sample'] = 'CC-2931'
CC2931_high = CC2931_gene_network[column].loc[CC2931_hi_exp]
CC2931_high['expression'] = 'high'
CC2931_high['sample'] = 'CC-2931'

centrality_plot = pd.concat([CC2344_low, CC2344_high, CC2931_low, CC2931_high], axis = 0)

#### Median values of centrality in the lowest and highest expressed metabolic genes ####
results = pd.DataFrame(index = column)
for i in column:
    results.at[i, 'CC2931_low_median'] = CC2931_low[i].median()
    results.at[i, 'CC2931_high_median'] = CC2931_high[i].median()
    results.at[i, 'CC2344_low_median'] =  CC2344_low[i].median()
    results.at[i, 'CC2344_high_median'] = CC2344_high[i].median()

#### Testing from differences in centrality between low and high expression metabolic genes  ####
for i in column:
    results.at[i, 'CC2344_fk'] = fligner(CC2344_high[i].dropna().values.tolist(), CC2344_low[i].dropna().values.tolist())[1]
    results.at[i, 'CC2931_fk'] = fligner(CC2931_high[i].dropna().values.tolist(), CC2931_low[i].dropna().values.tolist())[1]
    if results.at[i, 'CC2344_fk'] < 0.05:
        results.at[i, 'CC2344_bm_pval'] = brunnermunzel(CC2344_high[i].dropna().values.tolist(), CC2344_low[i].dropna().values.tolist(), alternative = 'greater')[1]
    else:
        results.at[i, 'CC2344_mw_pval'] = mannwhitneyu(CC2344_high[i].dropna().values.tolist(), CC2344_low[i].dropna().values.tolist(), alternative = 'greater')[1]
    if results.at[i, 'CC2931_fk'] < 0.05:
        results.at[i, 'CC2931_bm_pval'] = brunnermunzel(CC2931_high[i].dropna().values.tolist(), CC2931_low[i].dropna().values.tolist(), alternative = 'greater')[1]
    else:
        results.at[i, 'CC2931_mw_pval'] = mannwhitneyu(CC2931_high[i].dropna().values.tolist(), CC2931_low[i].dropna().values.tolist(), alternative = 'greater')[1]
    
#### Distribution of centrality values for metabolic genes with the highest and least expression level ####
for i in column: 
    plt.figure()
    sns.violinplot(data = centrality_plot, x = 'sample', y = i, hue = 'expression')
    plt.xlabel('Strains')
    plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left', title = 'Expression')
    plt.savefig('/research/projects/chlamydomonas/MAexpression/analysis/centrality/high_vs_low_expression_' + i + '.pdf', format = 'pdf', dpi = 150, bbox_inches = 'tight')
display(results)

for i in column: 
    plt.figure()
    sns.lineplot(data = centrality_plot, x = 'sample', y = i, hue = 'expression')
    plt.xlabel('Strains')
```
